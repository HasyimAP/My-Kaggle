# Import Libraries

```python
import time
import pandas as pd

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

# Initialize Selenium WebDriver

```python
CHROME_BINARY_LOCATION = "/usr/bin/chrome-linux64/chrome"
CHROMEDRIVER_BINARY_LOCATION = "/usr/bin/chromedriver-linux64/chromedriver"

def add_driver_options(options):
    """
    Add configurable options
    """
    chrome_options = Options()
    for opt in options:
        chrome_options.add_argument(opt)
    return chrome_options

def initialize_driver():
    """
    Initialize the web driver
    """
    driver_config = {
        "options": [
            "--headless",
            "--no-sandbox",
            "--start-fullscreen",
            "--allow-insecure-localhost",
            "--disable-dev-shm-usage",
            "user-agent=Chrome/116.0.5845.96"
        ],
    }
    options = add_driver_options(driver_config["options"])
    options.binary_location = CHROME_BINARY_LOCATION
    driver = webdriver.Chrome(
        executable_path=CHROMEDRIVER_BINARY_LOCATION,
        options=options)
    return driver
```

# Load All The Available Car

```python
# Initiate selenium webdriver
driver = initialize_driver()
driver.get("https://www.anwb.nl/auto/private-lease/anwb-private-lease/aanbod/")

while True:
    try:
        # Locate the button element
        button = driver.find_element(By.XPATH, '//*[@id="main-content"]/div/div[3]/div/div/div/div/div/section/div[2]/section/div[2]/div/div/div/button')
        
        # Check if the button is visible
        if button.is_displayed():
            # Click the button
            driver.execute_script('arguments[0].click();', button)
            
            # Wait for the page to load and the new data to appear
            time.sleep(10)
            
        else:
            # Break the loop if the button is no longer visible
            break
    
    except:
        # Break the loop if the button cannot be found
        break
```

```python
# Check the html page source
soup = BeautifulSoup(driver.page_source, 'html')
soup
```

# Get All The Data

```python
# Find all the car's name and group it accordingly
car_names = []
car_manufacturer = []
car_model = []

cars = soup.find_all('span', class_ ='sc-fQgSAe duLMLp sc-eODrEC cXeHSQ PONCHO-typography PONCHO-typography--button-link')
for car in cars:
    car = car.text
    car_names.append(car.strip())
    car = car.strip()
    car_manufacturer.append(car.split(' ', 1)[0])
    car_model.append(car.split(' ', 1)[1])
```

```python
# Get the car's price
car_prices = []

prices = soup.find_all('span', class_ ='sc-Gppvi kvInjJ PONCHO-typography PONCHO-typography--price-text-xl')
for price in prices:
    price = price.text
    car_prices.append(price.replace(u'\xa0', u' '))
```

```python
# Get all html elements that contain the link for detail of each car
car_elements = soup.find_all('a', class_ = 'sc-cDJyZ vNUGH')

# Initiate a new webdriver to open the link that contains the car's details
driver2 = initialize_driver()

# Collect and group all the needed data
car_fuel_type = []
car_chassis = []
car_link = []

for element in car_elements:
    car_link.append(element['href'])
    
    driver2.get(element['href'])
    soup2 = BeautifulSoup(driver2.page_source, 'html')
    
    if soup2.find('li', 'sc-27797e7b-0 TbkrF sc-f818400a-1 btllSG PONCHO-list__item PONCHO-typography PONCHO-typography--body-text'):
        list_items = soup2.find_all('li')
        fuel_types = [item for item in list_items if 'Brandstof' in item.text]

        for fuel_type in fuel_types:
            car_fuel_type.append(fuel_type.text.split(' ', 1)[1].strip())

        vehicle_chassis = [item for item in list_items if 'Carrosserie' in item.text]
        for chassis in vehicle_chassis:
            car_chassis.append(chassis.text.split(' ', 1)[1].strip())
        
    elif soup2.find('dl', class_ = 'sc-ePDLzJ cWOBop'):
        list_items = soup2.find_all('dd', class_ = 'sc-bXCLTC dhfvJu sc-iXzfSG jSQjhJ PONCHO-description-list__description PONCHO-typography PONCHO-typography--body-text')
        
        car_fuel_type.append(list_items[1].text)
        car_chassis.append(list_items[4].text) 
    else:
        car_fuel_type.append('-')
        car_chassis.append('-')
```

# Save The Data To Pandas DataFrame And Save As CSV

```python
df = pd.DataFrame({
    'Car Name': car_names,
    'Manufacturer': car_manufacturer,
    'Model': car_model,
    'Price': car_prices,
    'Fuel Type': car_fuel_type,
    'Vehicle Chassis': car_chassis,
    'Link': car_link
})

df
```

```python
df.to_csv('rental_car_data.csv', index=False)
```
